```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from google.colab import drive
drive.mount('/content/drive')

#Importing Cartera Data
dataset = '/content/drive/MyDrive/Kinetics/Rat_and_Rabbit_Processed_3rd_Crop_Final_27Feb23.xlsx'
dataset = pd.read_excel(dataset, sheet_name='Kinetic Tables')
dataset.drop(labels='Unnamed: 0', axis=1, inplace=True)
dataset.drop(2, inplace=True)
dataset.dropna(inplace=True)
dataset.columns.values[1]='Lot Number'
dataset.columns.values[3]='K_a'
dataset.columns.values[5]='K_d'
dataset.columns.values[8]='K_D'
for i in dataset['Lot Number']:
  dataset['Lot Number'] = dataset['Lot Number'].replace([i], i[-3:])

#Creating the Table for the Averages for Catera Data
avg = pd.pivot_table(data=dataset, index='Lot Number',
                     values=['K_a', 'K_d', 'K_D'], aggfunc = 'mean')

#Importing Lot Numbers
lot_number = '/content/drive/MyDrive/Kinetics/Lot_numbers.xlsx'
lot_number = pd.read_excel(lot_number, sheet_name='Sheet2', usecols=['Lot Number', 'Clone ID'])

#formating Lot Number
for i in lot_number['Lot Number']:
  lot_number['Lot Number'] = lot_number['Lot Number'].replace([i], i[-3:])
lot_number.set_index('Lot Number', inplace=True)


#Mergin and organizing the Two DataFrames
merge = pd.concat([avg, lot_number], axis=1)
merge.dropna(inplace=True)
merge=merge.reset_index().set_index('Clone ID')
merge=merge.loc[:,['Lot Number', 'K_a', 'K_d', 'K_D']]

#Saving the DataFrame
merge.to_csv('/content/drive/MyDrive/Carterra.csv')

display(merge)
```

```python
#Generating the Graphs
graph = merge.drop(labels='Lot Number', axis=1)
dict_e = {0: 'K_a Avg', 1: 'k_d Avg', 2:'k_D Avg'}

graph = merge.drop(labels='Lot Number', axis=1)

fig, axs = plt.subplots(len(graph.columns), figsize=(15,20))
fig.subplots_adjust(hspace=.6)

for i, j in dict_e.items():
  axs[i].bar(graph.index, graph.iloc[:, i])
  axs[i].title.set_text(j)
  axs[i].set_ylabel('Value')
  axs[i].set_xlabel('Clone ID')
  axs[i].tick_params(axis='x', labelrotation=60)

images_dir = '/content/drive/My Drive/'
plt.savefig(f"{images_dir}/Carterra_Graphs.png")

plt.show()
```
