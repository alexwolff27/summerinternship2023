```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from google.colab import drive
drive.mount('/content/drive')

#Importing the Excel File
dataset_1 = '/content/drive/MyDrive/Kinetics/Processed Rage Chimeric_21Jun23.csv'
dataset = pd.read_csv(dataset_1)

display(dataset)

#Creating the Table for the Averages
avg = pd.pivot_table(data=dataset, index='Name', aggfunc = 'mean')
avg.head()
```

```python
#Generating the Graphs
dict_e = {0: 'K_D Avg', 1: 'k_off Avg', 2:'k_on Avg'}

fig, axs = plt.subplots(len(avg.columns), figsize=(15,10))
fig.subplots_adjust(hspace=0.8)

for i, j in dict_e.items():
  axs[i].bar(avg.index, avg.iloc[:, i])
  axs[i].title.set_text(j)
  axs[i].set_ylabel('Value')
  axs[i].set_xlabel('Clone ID')
  axs[i].tick_params(axis='x', labelrotation=60)

plt.show()

```
